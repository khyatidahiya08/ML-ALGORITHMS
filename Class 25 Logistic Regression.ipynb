# LOGISTIC REGRESSION

# Logistic Regression is a statistical method for analyzing a data set in which there are one or more independent variables that determine an outcome, the outcome is measured with binary variable


#>-dependent variable is binary
#>-1(true,success),0(false,failure)
#>-goal is to find the best fitting model for independent  and dependent  variable relationship
#>-independent variable can be continuous or binary
#>-also called logistic regression
#>-used in machine learning
#>-deals with probability to measure the relation between dependent and independent variable

```python
df=pd.read_csv('hire_data1.csv')
```

```python
#logistic regression works on logic i.e in the form of probabilities (0 and 1)where 0 stands for no and 1 stands for yes
#example- during weather forecast the probability of rain goes from 0 to 1
```

```python
features=df.columns[:5]
```

```python
features#independent variables
```

```python
x=df[features]
x
```

```python
y=df[['Hired']]#dependent variable
y
```

```python
y=np.ravel(y)
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
print(metrics.accuracy_score(y_test,y_pred)*100)
#print(metrics.accuracy_score(y_test,y_pred.round())*100) - this statement also works in logistic regression
```

```python
lr.predict(np.array([80,0,1,60,60],dtype='float').reshape(1,-1))#prediction for multi x value
#here we entered the values of the independent x variables and predict whether the person is hired or not
#float is used because the x values that we enter might be float values
#1 is used because to have the output in form of a one dimensional array
#-1 is used so the values of x start from column 0
```
